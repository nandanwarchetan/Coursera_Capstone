```python
import io
import requests
```

```python
url="https://cocl.us/Geospatial_data"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
c.head()
```

```python
processed_df = pd.merge(left=unprocessed_df, right=c, left_on='Postal Code', right_on='Postal Code')
```

```python
processed_df.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(processed_df['Borough'].unique()),
        processed_df.shape[0]
    )
)
```

```python
processed_df.to_csv('toronto_neighborhoods.csv', index=False)
```
